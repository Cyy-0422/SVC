# 乳腺癌数据集下探索核函数的性质

## 探索kernel该如何选取

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer 
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
data 
```

```python
# 查看特征矩阵
X = data.data
X
```

```python
X.shape
```

```python
y = data.target
np.unique(y)
```

查看数据集的分布

但是由于特征太多，我们强行取第一个和第二个特征来查看一下数据大概的分布

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=420)
```

```python
kernel=['linear','poly', 'sigmoid','rbf']
for kernel in kernel:
    clf = SVC(kernel=kernel, gamma='auto', degree=1, cache_size=5000).fit(X_train, y_train) # 实例化SVC
    print("The accuracy under kernel %s is %f" % (kernel, clf.score(X_test,y_test)))
```

发现
>线性核函数的效果最好，而且当degree=1时，多项式核函数处理的就是线性问题，从结果来看，效果也不错，再结合之前可视化的结果，可以证明乳腺癌的数据集是一个线性可分的数据集

问题
>从之前探索核函数性质的时候实，我们了解到rbf比较万能，在线性数据集上也可以表现得很好，但是为什么结果却这么糟糕？


```python
# 描述性统计
X = pd.DataFrame(X)
X.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
```

1. 发现:
    * 量纲不统一(均值标准差相差较大)
    * 数据呈偏态分布（当量纲较大时，位于%99的数据与最大值相差很大）
>观察%99的数据和最大值之间的差距，若很大，则为右偏分布

    >观察%1的数据和最小值之间的差距，若很大，则为右偏分布
    
2. 解决：标准化

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X) 
X = pd.DataFrame(X)
X.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=420)
kernel=['linear','poly', 'sigmoid','rbf']
for kernel in kernel:
    clf = SVC(kernel=kernel, gamma='auto', degree=1, cache_size=5000).fit(X_train, y_train)
    print("The accuracy under kernel %s is %f" % (kernel, clf.score(X_test,y_test)))
```

结论
>标准化之后，四个核函数的效果都变得比较好，sigmoid核函数都达到了%95的准确率。

经验
>用SVM分类时，先看分布，如果量纲不统一，则进行标准化


## 选取与核函数相关的参数

### 高斯核函数
通过画学习曲线来找最优的gamma值


```python
score=[]
gamma_range = np.logspace(-10, 1, 50) # 返回在对数刻度上均匀间隔的数字
for i in gamma_range:
    clf=SVC(kernel="rbf",gamma=i,cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
print(max(score), gamma_range[score.index(max(score))])
plt.plot(gamma_range,score)
plt.show()
```

```python
np.logspace(-10, 1, 50) 
```

### 多项式核函数


```python
from sklearn.model_selection import GridSearchCV # 带交叉验证的网格搜索
from sklearn.model_selection import StratifiedShuffleSplit # 设置交叉验证的模式

gamma_range = np.logspace(-10,1,20)
coef0_range = np.linspace(0,5,10)
param_grid = dict(gamma=gamma_range, coef0=coef0_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=420)
grid = GridSearchCV(SVC(kernel="poly",degree=1,cache_size=5000),
                    param_grid=param_grid, cv=cv)
grid.fit(X, y)
print("The best parameters are %s with a score of %0.5f" % (grid.best_params_,
grid.best_score_))
```

# 软间隔与重要参数C

```python
score = []
C_range = np.linspace(0.01,30,50)
for i in C_range: 
    clf=SVC(kernel="linear", C=i, cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
    
print(max(score), C_range[score.index(max(score))])
plt.plot(C_range,score)
plt.show()
```

```python
score = []
C_range = np.linspace(0.01,30,50)
for i in C_range: 
    clf=SVC(kernel="rbf", C=i, cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
    
print(max(score), C_range[score.index(max(score))])
plt.plot(C_range,score)
plt.show()
```

```python
# 进一步细化
score = []
C_range = np.linspace(24,26,50)
for i in C_range: 
    clf=SVC(kernel="rbf", C=i, cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
    
print(max(score), C_range[score.index(max(score))])
plt.plot(C_range,score)
plt.show()
```

我们找到了乳腺癌数据集上的最优解：rbf核函数下的98.83%的准确率。
